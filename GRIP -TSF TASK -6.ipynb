*Importing libraries in Python*

```python
import sklearn.datasets as datasets
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

**Loading the iris dataset**

```python
iris=datasets.load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
y=iris.target
print(y)
```

**split data intotrain and test set**

```python
from sklearn.model_selection import train_test_split
df_train,df_test,y_train,y_test = train_test_split(df,y,test_size = 0.40,random_state = 42)
```

```python
print(df_train.shape)
print(df_test.shape)
print(y_train.shape)
print(y_test.shape)
```

**defining the decision Tree algorithm**

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df_train,y_train)

print('Decision Tree Classifer Created')
```

```python
y_pred = dtree.predict(df_test)
```

```python
y_pred
```

**confusion matrix**

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn import tree
```

**Visualizing the Decision Tree**

```python
fn = ['sepal length(cm)','sepal width(cm)','petal lenght(cm)','petal width(cm)']
cn = ['setosa','versicolour','verginica']

fig, axes = plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi =350)

tree.plot_tree(dtree, feature_names =fn, class_names=cn, filled=True);
```
